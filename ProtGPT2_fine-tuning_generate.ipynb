```python
start = time.time()

# 生成したいシーケンスの総数を定義
total_sequences = 1000000

# 一度に生成するシーケンスの数を定義
batch_size = 500

# 生成したい配列数
seq_num = 1000

# 生成したシーケンスを保存するリストを作成
generated_sequences = []

sequences = []
ppls = []
lengths = []

# pplを計算するための関数
def calculatePerplexity(sequence, model, tokenizer):
    input_ids = torch.tensor(tokenizer.encode(sequence)).unsqueeze(0) 
    input_ids = input_ids
    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)
    loss, logits = outputs[:2]
    return math.exp(loss)

# total_sequences/batch_size 回のイテレーションを行い、シーケンスを生成
for _ in range(total_sequences // batch_size):
    # model.generateでシーケンスを生成します
    outputs = model.generate(
        input_ids, 
        top_k=950, 
        repetition_penalty=1.2,
        max_length=200,
        eos_token_id=0,
        do_sample=True,
        num_return_sequences=batch_size)
    
    # 生成したシーケンスをCPUメモリに移動し、リストに追加（GPUメモリを開放）
#     generated_sequences.extend(outputs.cpu().tolist())

    # 配列を1件ずつ取り出してsequence長とpplを計算
    for seq in outputs:
        sequence = tokenizer.decode(seq, skip_special_tokens=True)
        ppl = calculatePerplexity(sequence, model, tokenizer)
        length = len(sequence)
        
#         ぞれぞれの条件を指定
        if (ppl <= 26) and (length <= 278) and (length >= 104) and (sequence[:2] == '\nM'):
            sequences.append(sequence)
            ppls.append(ppl)
            lengths.append(length)
            
    outputs = []
            
    if len(lengths) >= seq_num:
        print(str(len(lengths)) + '件生成されました。')
        break

e_time = time.time() - start
print ("e_time:{0}".format(e_time) + "[s]")


# ここからファイル保存（csvとfastaで保存）
a = pd.DataFrame({'sequence': sequences,'ppl': ppls, 'length':lengths})
a['sequence'] = a['sequence'].str.replace('\n', '')
a['length'] = a['sequence'].map(lambda x: len(x))
a.to_csv("generated_sequence_small_epoch50_1000.csv", index=False)

def write_fasta(sequences, file_name):
    with open(file_name, 'w') as file:
        for i, seq in enumerate(sequences):
            header = f">seq_{i+1}"
            file.write(header + "\n")
            file.write(seq + "\n")

# 例のリスト（配列）
sequences = a['sequence'].to_list()
# ファイルに書き出す
write_fasta(sequences, 'generated_sequence_small_epoch50_1000.fasta')
```
